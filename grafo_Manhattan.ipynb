```python
import networkx as nx
G = nx.Graph()

#create nodes
G.add_node("Washington Heights") #Washington Heights
G.add_node("Harlem") #Harlem
G.add_node("Upper Manhattan") #Upper Manhattan
G.add_node("Midtown Manhattan") #Midtown Manhattan
G.add_node("Lower Manhattan") #Lower Manhattan

print(G.nodes())

#create edges
G.add_edge("Washington Heights", "Harlem")
G.add_edge("Harlem","Upper Manhattan")
G.add_edge("Upper Manhattan","Midtown Manhattan")
G.add_edge("Midtown Manhattan","Lower Manhattan")

print(G.edges())

#draw graph
nx.draw(G, with_labels = True, node_size = 1000, edge_color= "red")
```
